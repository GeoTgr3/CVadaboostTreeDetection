```python

def adaboos(data): # it's a function which is defined by "def" abd the function name is "adaboos" and it has only one argument which is "data".
    # data = pd.read_csv("res.csv") # pd.read_csv() is used for read CSV file.
    y = data[data.columns[-1]] # data.columns[-1] is used for read the last column of the dataset. -1 means it take the last column. after it store the last column value within "y" variable.
    print(y) # it's print the "y" value.
    X = data.drop(data.columns[-1], axis=1) ## drop is used for remove a column or row from the dataset.
     ##here use "axis = 1" which means its drop column from the dataset if you drop row then you have to use "axis = 0".
    ##"data.columns[-1] " means the last column.
    ##"data.drop(data.columns[-1], axis=1)" means drop the the last column from the dataset and store the rest of the column within "X" variable."""
    print(X) # print the x's value
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4) ## 'train_test_split' it's split the input data and target data
    ##here the dataset X is split for X_train and X_test also split the target data between y_train and y_test.
    ##wintin 'train_test_split()' use test_size=0.4 means 40% data will be used for test and 60% data use for train.   """
    print(y_train) # it's mean you see the whole value of y_train data


    clf = AdaBoostClassifier(n_estimators=100, random_state=0)# it's a classification algorithm. n_estimators=100 means it's use 100 decision tree and  AdaBoostClassifier take the average feature importance from 100 decision tree.
    # random_state = 0 means it's take train input data randomly not a sequence of index. and store the pattern in clf variable.

    model = clf.fit(X_train, y_train) # clf.fit() use for fit the X_train and y_train data with the pattern of AdaBoostClassifier
    y_pred = model.predict(X_test) # model.predict() predict the X_test value from the previous experienced value.
    print("Accuracy: ", metrics.accuracy_score(y_test, y_pred)) # metrics.accuracy_score measure the performance of this algorithm. it's measure the loss between actual value and predicted value.


```

```python

def dataGenerator(image, image_LAB, circle_centers, diam): # it's a function which is defined by "def" abd the function name is "dataGenerator" and it has four argument which are "image, image_LAB, circle_centers, diam".
    circle_centers_df = pd.read_csv(circle_centers, header=None) # pd.read_csv() is used for read CSV file. "circle_centers" is the file name and header = None means the first row of the csv file will be treated as column names.
    rows, cols, _ = image.shape # it's give how much row and columns have in dataset. total row is stores in rows and totals columns are store in cols variable. and the (_) is used ignore the value of channels.
    print(rows, " ", cols) # print rows and columns value
    df_result = pd.DataFrame(0, np.arange(rows * cols), np.arange(7)) # here pd.DataFrame() arguments are "0,np.arange(rows * cols), np.arange(7)" . 0 is data in dataframe,np.arange(rows * cols) is the number of index and  np.arange(7) is the number of columns.
    rows_circles, cols_circles = circle_centers_df.shape # # it's give how much row and columns have in dataset. total row is stores in rows_circles and totals columns are store in cols_circles variable.
    for i in range(rows_circles): # for loop use for iterate the program. and the program continue total number of rows_circles. range means the duration for how much time the for loop will be continue.
        for j in range(cols_circles): #the program continue total number of cols_circles. range means the duration for how much time the for loop will be continue.
            elements = tuple(circle_centers_df.iloc[i, j][1:-1].split(", ")) # index location of circle_centers_df[i,j] from row 1 to the last row split all data where have comma(,) and store those data within a tuple. split(", ") means separeted data where have comma
            c_i = int(elements[0]) # elements[0] take the first value from the tuple and it convert the value to integer using int()
            c_j = int(elements[1]) # elements[1] take the second value from the tuple and it convert the value to integer using int()

            for temp_i in range(c_i - diam // 2, c_i + diam //2): ## in python '//' this sign use for floor division that have to integer value such as 5//2 = 2
            ##loop continue from c_i minus diam //2 value to c_i + diam //2 """
                for temp_j in range(c_j - diam // 2, c_j + diam //2):
                    if temp_i >= 0 and temp_i < rows and temp_j >= 0 and temp_j < cols: # this is the condition line. if the condition is true then program into the next condition. if the condion false then it back to the loop and again chaeck the condition.this work continue umtil the range end.
                        if (c_i - temp_i) ** 2 + (c_j - temp_j) ** 2 <= (diam//2) ** 2: # it's check the condition
                            df_result.iloc[temp_i * cols + temp_j, 6] = 1 # when the condition is true then it's take the value 1 in this 'df_result.iloc[temp_i * cols + temp_j, 6]' index location.

    for i in range (rows):
        for j in range(cols):
            [B, G, R] = image[i, j] # each image has 3 colors RGB. Here image[i, j] means a pixel within image. and each pixel represent any color within RGB and the color value is stored in [B,G,R] variable.
            df_result.iloc[i * cols + j, 0] = B # the index location of '[i * cols + j, 0]' the color is B(Blue). first iterate the i = 0, cols = total number of columns that find out above. j = 0, so the row value is 0*cols+0 and the column is 0.
            df_result.iloc[i * cols + j, 1] = G
            df_result.iloc[i * cols + j, 2] = R
            [L, a, b] = image_LAB[i,j]
            df_result.iloc[i * cols + j, 3] = L
            df_result.iloc[i * cols + j, 4] = a
            df_result.iloc[i * cols + j, 5] = b
    df_result.to_csv("res.csv", index=False) # here convert the all index_location value is converter to CSV files and the file name is 'res.csv', index = False means you ignore the index value;

    return df_result # it's return the dataset to the calling function


```

```python

def template_drawing(original_image, df, diam): # it's a function which is defined by "def" abd the function name is "template_drawing" and it has 3 argument which are "original_image, df, diam".

    theta_corr=0.65 # user defined variable that you can vary
    overlap_neighors=5
    image=original_image # here orginal_image copy within image variable

    rows, columns= df.shape # find out how much row and column have. shape mainly determine how much dimension have of the dataset
    thickness=1
    # center_coordinates=(100,100)
    radius=int(diam/2) # its divide the diam by 2 and convert to integer value which is radius
    color = (255, 0, 0) # color where only one color is high that is 255, and others color is low.
    step=int(diam//4) # calculate the floor value of diam by 4 and convert to integer value.
    tree_log=[] # take a empty list for store data later
    center_list_row=[] # take a empty list for store data later
    print(" drawing the templates")
    for i in range(0,rows-overlap_neighors,overlap_neighors+1): # continue the loop from 0 to rows-overlap_neighors and the step will be overlap_neighors+1 from one step to the next step. ex: if range(0,5,2) then the output will be (0, 2, 4)
        for j in range(0, columns-overlap_neighors, overlap_neighors+1): #range(strat from 0 value, continue till columns-overlap_neighors and step from one j value to next j value is overlap_neighors+1)
            best_score=df.iloc[i,j]
            c_i = step * (i) + diam / 2
            c_j = step * (j) + diam / 2
            center_coordinates = (int(c_j), int(c_i)) # it's determine the coordinate, example like (0,0),(5,8), etc.
            for i_sqr in range(overlap_neighors):
                for j_sqr in range(overlap_neighors):
                    #check the condition if the score is above the theta_corr and if it is maximum between the neighbors
                    if df.iloc[i+i_sqr,j+j_sqr]>theta_corr: # if the condition is true then into the next if condition  otherwise go to the else condion

                        if df.iloc[i+i_sqr,j+j_sqr]>best_score:
                            df.iloc[i,j]=0
                            best_score=df.iloc[i+i_sqr,j+j_sqr]
                            c_i = step * (i + i_sqr) + diam / 2
                            c_j = step * (j + j_sqr) + diam / 2
                            center_coordinates = (int(c_j), int(c_i))
                        else:
                            df.iloc[i + i_sqr, j + j_sqr]=0
                    else:
                        df.iloc[i + i_sqr, j + j_sqr] = 0

            if best_score>theta_corr: # if the best_score is greater than theta_corr then go to the next line within the condition
                image = cv2.circle(image, center_coordinates, radius, color, thickness) # draw a circle using cv2.circle() the arguments within cv2.circle() is "image" which is the main image, center_coordinates means circle draw from this position and radius means the field of the circle, color means what will be the color of the circle and thickness determine the outer line of the circle.

                center_list_row.append(center_coordinates) # store the center_coordinates value within 'center_list_row' using append

        #some operations here for the center_list_row to find the line
        # image=cv2.line(image, center_list_row[0],center_list_row[-1],color, thickness)
        tree_log.append(center_list_row) # store the 'center_list_row' value within 'tree_log' using append
        center_list_row=[] # take an empty list for store data



    return df , image, tree_log # finaly return these values to the calling function.
    # print(df)
```

```python

def localization(blk_img, diam, scale=20): # the function name is 'localization' and here are 3 arguments
   # from offline calculations we found out that an image with scale 1:20 has 224 meter width
   # Each pixel is 0.23 m
   # Our assumption is that each tree contains 16 to 24 pixels in the transfered image
    height, width, _ = blk_img.shape # image is 3 dimensional. it's height, weight and channels. (_) underscore sign means ignore the chammels
    dataframe_list = [] # take an empty list
    print("Template matching for diam="+str(diam)) # print dial value in a string data type so congvert diam dtype to string using 'str()'
    step = int(diam / 4) # step take from diam,  diam divided by 4 and convert to integer
    num_circles_row = (height-diam)// step+1 # num_circles_row will be (height - diam) divided by (step + 1), here take (//) sign for take floor integer value
    num_circles_col = (width-diam)// step+1 # num_circles_row will be (weight - diam) divided by (step + 1), here take (//) sign for take floor integer value
    df = pd.DataFrame(0, np.arange(num_circles_row), np.arange(num_circles_col))
    df_binary=df # data copy from df variable to a new variable
    print("number of templates along the rows and columns:")
    print(num_circles_row,num_circles_col)
    for i in range(num_circles_row):
        for j in range(num_circles_col):
            c_i=step*i+diam/2
            c_j=step*j+diam/2
            total_pixels=0
            white_pixels=0


            for x in range (int(c_i-diam/2),int(c_i+1+diam/2)-1):
                for y in range (int(c_j-diam/2),int(c_j+1+diam/2)-1):
                    # print("x",x,"y",y)
                    if ((x - c_i) ** 2 + (y - c_j) ** 2 <= (diam / 2) ** 2): # here is the condition, (x - c_i) ** 2 means (x subtraction from c_i) power 2 add (y subtraction from c_j) power 2 less than or equal with "(diam / 2) ** 2" then the condition will be True. here, (**) is the power sign
                        if (blk_img[x,y]==[255,255,255]).all():
                            white_pixels+=1 # it's doing add 1 with white_pixels. like "white_pixels = white_pixels + 1"

                        total_pixels+=1 # it's doing add 1 with total_pixels. like "total_pixels = total_pixels + 1"
            # print(total_pixels)

            template_score=white_pixels/total_pixels # divide white_pixels by total_pixels
            # print("i"+str(i)+"j"+str(j))
            # print(template_score)

            df.iloc[i,j]=template_score # assign the value of template_score to [i, j] index location within df dataset


    # print(df)
    return df # return back to the calling function
```

```python

def masking(img_name):
    # construct the argument parse and parse the arguments
    ap = argparse.ArgumentParser() #initialijes the parser so that you can start to add custom arguments.
    ap.add_argument("-i", "--image", help = img_name) # add the custom arguments.
#   args = vars(ap.parse_args())
    # load the image
    image = cv2.imread(img_name) # read image using cv2 library and take the image file name within cv2.imread(). here img_name is the image name
    image2=image.copy() # copy the image and store it to the image2 variable

    height, width,_=image.shape # take the shape of the image. here only take height and weight of an image
    print(height, width) # using print function you can see the value of height and weight
    upper=[86,255,255] # it's a thresehold value, you can change the value if you want.
    lower=[36,0,0] # it's also a thresehold value, you can change the value if you want.
    # print(image[50,50])
    # cv2.imshow('pseudocolor',image)
#   cv2.waitKey(0)

    for i in range(0, height):
        for j in range(0,width):
            if (image2[i,j]<=upper).all() and (image2[i,j]>=lower).all(): # if both condition is true then execute next line of codes. '.all()' means all value have to True
                image2[i, j]= [255, 255, 255] # in image1, the coordinate of image2[i,j] all values are 255
            else:
                image2[i, j]=[0,0,0] # in image1, the coordinate of image2[i,j] all values are 0
    return image, image2 # return image and image2 value to the calling function
```
